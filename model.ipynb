# **Breast Cancer Prediction using Logistic regression**

```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collection & Processing

```python
#loading the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
# loading data to pd dataframe
df = pd.DataFrame(breast_cancer_dataset.data,columns=breast_cancer_dataset.feature_names)
```

```python
df.head(3)
```

```python
df['target'] = breast_cancer_dataset.target
```

```python
df.head(3)
```

```python
df['target'].value_counts()
```

```python
df.info()
```

```python
# since no categorical value so no need of one hot encoding
df.groupby('target').mean()
```

## 1 --> Benign

## 0 --> Malignant

```python
X = df.drop('target',axis='columns')
X.head()
```

```python
Y = df['target']
Y.head(2)
```

### Splitting data into Test And Train data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
print(X_train.shape,X_test.shape)
```

```python
model = LogisticRegression(max_iter=3000) # increased the number of iterations
model.fit(X_train,Y_train)
```

```python
model.score(X_test,Y_test)
```

### **Using K fold cross validation method to check the accuracy of the logistic regression model**

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv = ShuffleSplit(n_splits=5,test_size=0.2,random_state=2)
scores = cross_val_score(model,X,Y,cv=cv)
print(scores)
np.mean(scores)
```
